```python
# BeautifulSoup Ìå®ÌÇ§ÏßÄ Î∂àÎü¨Ïò§Í∏∞
from bs4 import BeautifulSoup as bs

# HTML ÏΩîÎìú ÏòàÏãú
html_text = """
<html>
  <body>
    <p class="greeting">ÏïàÎÖïÌïòÏÑ∏Ïöî</p>
    <p class="greeting">Î∞òÍ∞ëÏäµÎãàÎã§</p>d
    <div id="message">Ïó¨Í∏∞ÏóêÎäî Î©îÏãúÏßÄÍ∞Ä ÏûàÏäµÎãàÎã§.</div>
    <div id="info">Ïó¨Í∏∞ÏóêÎäî Ï†ïÎ≥¥Í∞Ä ÏûàÏäµÎãàÎã§.</div>
  </body>
</html>
"""

# HTML ÌååÏã±
soup = bs(html_text, 'html.parser')
```

```python
# BeautifulSoup Ìå®ÌÇ§ÏßÄ Î∂àÎü¨Ïò§Í∏∞
# Ï£ºÎ°ú bsÎ°ú Ïù¥Î¶ÑÏùÑ Í∞ÑÎã®Ìûà ÎßåÎì§Ïñ¥ÏÑú ÏÇ¨Ïö©Ìï®
from bs4 import BeautifulSoup as bs  # BeautifulSoupÎ•º bsÎùºÎäî Ïù¥Î¶ÑÏúºÎ°ú Í∞ÄÏ†∏Ïò¥

# htmlÏùÑ Ïûò Ï†ïÎ¶¨Îêú ÌòïÌÉúÎ°ú Î≥ÄÌôò
html = bs(html_text, 'html.parser')  # html_textÎ•º ÌååÏã±ÌïòÏó¨ BeautifulSoup Í∞ùÏ≤¥Î°ú Î≥ÄÌôò
# 'html.parser'Îäî HTMLÏùÑ Ï≤òÎ¶¨ÌïòÎäî ÌååÏÑúÎ°ú, ÌååÏù¥Ïç¨ ÎÇ¥Ïû• HTML ÌååÏÑúÎ•º ÏÇ¨Ïö©

# Í≤∞Í≥º Ï∂úÎ†•
print(html)  # ÌååÏã±Îêú htmlÏùÑ Ï∂úÎ†•
```

```python
# find() - Ï≤´ Î≤àÏß∏ <p> ÌÉúÍ∑∏Î•º Ï∞æÍ∏∞
print("find() Í≤∞Í≥º:")
print(soup.find('p'))  # Ï≤´ Î≤àÏß∏ <p> ÌÉúÍ∑∏
```

```python
# find_all() - Î™®Îì† <p> ÌÉúÍ∑∏Î•º Ï∞æÍ∏∞(Î¶¨Ïä§Ìä∏ ÌòïÌÉú)
print("find_all() Í≤∞Í≥º Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú Ï∂úÎ†•:")
print(soup.find_all('p')) # Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú Ï∂úÎ†•
```

```python
# find_all() - Î™®Îì† <p> ÌÉúÍ∑∏Î•º Ï∞æÍ∏∞
print("find_all() Í≤∞Í≥º ÌïòÎÇòÏî© Ï∂úÎ†•:")
p_tags = soup.find_all('p')  # Î™®Îì† <p> ÌÉúÍ∑∏ ÌïòÎÇòÏî© Ï∂úÎ†•
for p in p_tags:
    print(p)
```

```python
# find() - class="greeting"Ïù∏ Ï≤´ Î≤àÏß∏ ÌÉúÍ∑∏ Ï∞æÍ∏∞
print("class='greeting'Ïù∏ Ï≤´ Î≤àÏß∏ ÌÉúÍ∑∏:")
print(soup.find(class_='greeting'))  # class="greeting"Ïù∏ Ï≤´ Î≤àÏß∏ ÌÉúÍ∑∏
```

```python
# find_all() - class="greeting"Ïù∏ Î™®Îì† ÌÉúÍ∑∏ Ï∞æÍ∏∞
print("class='greeting'Ïù∏ Î™®Îì† ÌÉúÍ∑∏:")
greeting_p_tags = soup.find_all(class_='greeting')  # class="greeting"Ïù∏ Î™®Îì† ÌÉúÍ∑∏
for tag in greeting_p_tags:
    print(tag)
```

```python
# find() - id="message"Ïù∏ Ï≤´ Î≤àÏß∏ ÌÉúÍ∑∏ Ï∞æÍ∏∞
print("id='message'Ïù∏ Ï≤´ Î≤àÏß∏ ÌÉúÍ∑∏:")
print(soup.find(id='message'))  # id="message"Ïù∏ Ï≤´ Î≤àÏß∏ ÌÉúÍ∑∏
```

```python
# find_all() - idÍ∞Ä ÏûàÎäî Î™®Îì† ÌÉúÍ∑∏ Ï∞æÍ∏∞
print("idÍ∞Ä ÏûàÎäî Î™®Îì† ÌÉúÍ∑∏:")
div_tags_id = soup.find_all('div', id=True)  # idÍ∞Ä ÏûàÎäî Î™®Îì† ÌÉúÍ∑∏
for tag in div_tags_id:
    print(tag)
```

```python
# requests Ìå®ÌÇ§ÏßÄ Í∞ÄÏ†∏Ïò§Í∏∞
import requests

# Í∞ÄÏ†∏Ïò¨ url Î¨∏ÏûêÏó¥Î°ú ÏûÖÎ†•
url = 'https://news.naver.com/'

# requestsÏùò getÌï®ÏàòÎ•º Ïù¥Ïö©Ìï¥ Ìï¥Îãπ urlÎ°úÎ∂ÄÌÑ∞ htmlÏù¥ Îã¥Í∏¥ ÏûêÎ£åÎ•º Î∞õÏïÑÏò§Í∏∞
response = requests.get(url)

# Ïö∞Î¶¨Í∞Ä ÏñªÍ≥†Ïûê ÌïòÎäî html Î¨∏ÏÑúÎ•º Î≥ÄÏàòÏóê Îã¥Í∏∞
html_text = response.text

# BeautifulSoup Ìå®ÌÇ§ÏßÄ Î∂àÎü¨Ïò§Í∏∞
from bs4 import BeautifulSoup as bs  # BeautifulSoupÎ•º bsÎùºÎäî Ïù¥Î¶ÑÏúºÎ°ú Í∞ÄÏ†∏Ïò¥

# htmlÏùÑ Ïûò Ï†ïÎ¶¨Îêú ÌòïÌÉúÎ°ú Î≥ÄÌôò
html = bs(html_text, 'html.parser')  # html_textÎ•º ÌååÏã±ÌïòÏó¨ BeautifulSoup Í∞ùÏ≤¥Î°ú Î≥ÄÌôò

# Ï£ºÏöî Îâ¥Ïä§ Ï†úÎ™©ÏùÑ Ï∞æÍ∏∞ ÏúÑÌïú ÏÑ†ÌÉùÏûê
titles = html.select('.cnf_news_title')

# Í∞Å Ï†úÎ™© Ï∂úÎ†•
if titles:
    for title in titles:
        print(title.get_text())
else:
    print("Îâ¥Ïä§ Ï†úÎ™©ÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")
    
# Ï†úÎ™© ÌïòÎÇò Ï∂úÎ†•
print()
title = html.select_one('.cnf_news_title')
print(title)
```
