```python
# requests 패키지 가져오기
import requests

# 가져올 url 문자열로 입력
url = 'https://www.naver.com'

# requests의 get 함수를 이용해 URL로 요청을 보내고, 해당 URL로부터 HTML 데이터를 포함한 응답을 받아오기
response = requests.get(url)

# 우리가 얻고자 하는 html 문서를 변수에 담기
html_text = response.text

# html 문서 출력
print(html_text)
```

```python
# BeautifulSoup 패키지 불러오기
from bs4 import BeautifulSoup as bs

# HTML 코드 예시
html_text = """
<html>
  <body>
    <p class="greeting">안녕하세요</p>
    <p class="greeting">반갑습니다</p>d
    <div id="message">여기에는 메시지가 있습니다.</div>
    <div id="info">여기에는 정보가 있습니다.</div>
  </body>
</html>
"""

# HTML 파싱
soup = bs(html_text, 'html.parser')
```

```python
# BeautifulSoup 패키지 불러오기
# 주로 bs로 이름을 간단히 만들어서 사용함
from bs4 import BeautifulSoup as bs  # BeautifulSoup를 bs라는 이름으로 가져옴

# html을 잘 정리된 형태로 변환
html = bs(html_text, 'html.parser')  # html_text를 파싱하여 BeautifulSoup 객체로 변환
# 'html.parser'는 HTML을 처리하는 파서로, 파이썬 내장 HTML 파서를 사용

# 결과 출력
print(html)  # 파싱된 html을 출력
```

```python
# find() - 첫 번째 <p> 태그를 찾기
print("find() 결과:")
print(soup.find('p'))  # 첫 번째 <p> 태그
```

```python
# find_all() - 모든 <p> 태그를 찾기(리스트 형태)
print("find_all() 결과 리스트 형태로 출력:")
print(soup.find_all('p')) # 리스트 형태로 출력
```

```python
# find_all() - 모든 <p> 태그를 찾기
print("find_all() 결과 하나씩 출력:")
p_tags = soup.find_all('p')  # 모든 <p> 태그 하나씩 출력
for p in p_tags:
    print(p)
```

```python
# find() - class="greeting"인 첫 번째 태그 찾기
print("class='greeting'인 첫 번째 태그:")
print(soup.find(class_='greeting'))  # class="greeting"인 첫 번째 태그
```

```python
# find_all() - class="greeting"인 모든 태그 찾기
print("class='greeting'인 모든 태그:")
greeting_p_tags = soup.find_all(class_='greeting')  # class="greeting"인 모든 태그
for tag in greeting_p_tags:
    print(tag)
```

```python
# find() - id="message"인 첫 번째 태그 찾기
print("id='message'인 첫 번째 태그:")
print(soup.find(id='message'))  # id="message"인 첫 번째 태그
```

```python
# find_all() - id가 있는 모든 태그 찾기
print("id가 있는 모든 태그:")
div_tags_id = soup.find_all('div', id=True)  # id가 있는 모든 태그
for tag in div_tags_id:
    print(tag)
```

```python
# requests 패키지 가져오기
import requests

# 가져올 url 문자열로 입력
url = 'https://news.naver.com/'

# requests의 get함수를 이용해 해당 url로부터 html이 담긴 자료를 받아오기
response = requests.get(url)

# 우리가 얻고자 하는 html 문서를 변수에 담기
html_text = response.text

# BeautifulSoup 패키지 불러오기
from bs4 import BeautifulSoup as bs  # BeautifulSoup를 bs라는 이름으로 가져옴

# html을 잘 정리된 형태로 변환
html = bs(html_text, 'html.parser')  # html_text를 파싱하여 BeautifulSoup 객체로 변환

# 주요 뉴스 제목을 찾기 위한 선택자
titles = html.select('.cnf_news_title')

# 각 제목 출력
if titles:
    for title in titles:
        print(title.get_text())
else:
    print("뉴스 제목을 찾을 수 없습니다.")
    
# 제목 하나 출력
print()
title = html.select_one('.cnf_news_title')
print(title)
```
