# 정적 웹 크롤링 연습

```python
# 미국 환율만 가지고 오기
import requests

# 네이버 금융 환율 정보가 있는 URL
url = "https://finance.naver.com/marketindex/exchangeList.naver"
response = requests.get(url)
html_text = response.text # HTML 내용을 텍스트 형식으로 가져오기
print(html_text)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(html_text, 'html.parser') # HTML 텍스트를 BeautifulSoup 객체로 파싱
print(soup)
```

```python
items = soup.select(".tbl_exchange tbody tr") # class="tbl_exchange"인 테이블의 <tbody> 아래에 있는 모든 <tr> 태그를 선택
exchange_rate = dict() # 비어있는 dict 생성

# 선택된 각 행을 순회하면서 필요한 데이터를 추출
for item in items: # 선택된 <tr> 태그(행)를 하나씩 처리
  title = item.select_one(".tit") # tit인 태그를 선택하여, 환율 이름가져오기
  title = title.get_text().strip() # get_text()를 사용해 텍스트만 추출하고, strip()으로 앞뒤 공백을 제거
  
  if title.find("USD") != -1: # 제목에 "USD"가 포함된 경우만 처리
    tds = item.select("td")  # 해당 행에 있는 모든 <td> 태그를 리스트로 가져오기
    # 각 <td>에서 환율 값을 추출
    sale = tds[1].get_text().strip()
    buy = tds[2].get_text().strip()
    sell = tds[3].get_text().strip()
    # 추출한 환율 데이터를 딕셔너리에 저장
    exchange_rate['sale'] = sale
    exchange_rate['buy'] = buy
    exchange_rate['sell'] = sell
    
# 최종적으로 추출된 환율 데이터를 출력   
print(exchange_rate)
```
